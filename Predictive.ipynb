```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('BDA_Data.csv')
data.head()
```

```python
data.describe()
```

```python
features = ['Open', 'High', 'Low', 'Volume']
```

```python
data[features] = (data[features] - data[features].mean()) / data[features].std()
```

```python
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[features])
```

```python
cluster_statistics = data.groupby('Cluster')[features].mean()
cluster_statistics
```

```python
label_mapping = {0: 'Sell', 1: 'Hold', 2: 'Buy'}
data['Label'] = data['Cluster'].map(label_mapping)
```

```python
company_labels = data.groupby('Company')['Label'].agg(lambda x: x.value_counts().idxmax()).reset_index()
res = pd.DataFrame(company_labels)
res
```
