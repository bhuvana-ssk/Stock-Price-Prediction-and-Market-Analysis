```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
data = pd.read_csv('BDA_Data.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['IsAboveThreshold'] = data['Close'] > 0.0
```

```python
basket = data.pivot_table(index='Date', columns='Company', values='IsAboveThreshold', fill_value=0)

min_support = 0.1  
frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)

min_confidence = 0.5  
association_rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=min_confidence)
```

```python
print("Frequent Itemsets:")
fi = pd.DataFrame(frequent_itemsets)
fi
```

```python

print("\nAssociation Rules:")
ar = pd.DataFrame(association_rules)
ar
```
